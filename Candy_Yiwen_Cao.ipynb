# Random Forest

```python
from sklearn.ensemble import RandomForestClassifier
from sklearn.model_selection import cross_val_score, KFold, cross_val_predict

from sklearn import metrics
from sklearn.metrics import accuracy_score, roc_curve, roc_auc_score, f1_score, classification_report, confusion_matrix

# Data sampling
remove=["welcomed","high-price","fruity","bar"] #remove high related features and variable
X = data1.drop(columns=remove)
y = data1.welcomed
```

```python
def cross_validation_classifer(classifer, X, y):
    # cross-validation because of small size data
    kfold = KFold(n_splits=5, shuffle=True, random_state=42)
    
    feature_importances = []
    cross_val_scores = []
    confusion_matrices = []
    roc_auc_scores = []
    classification_reports = []
    
    for train_index, test_index in kfold.split(X):
        X_train, X_test = X.iloc[train_index], X.iloc[test_index]
        y_train, y_test = y.iloc[train_index], y.iloc[test_index]
        
        classifer.fit(X_train, y_train)
        
        # feature_importance to find the drivers for welcomed candy
        feature_importances.append(classifer.feature_importances_)
        
        y_pred = classifer.predict(X_test)
        
        cross_val_scores.append(classifer.score(X_test, y_test))
        confusion_matrices.append(confusion_matrix(y_test, y_pred))
        
        report = classification_report(y_test, y_pred)
        classification_reports.append(report)
        
        
        roc_auc_scores.append(roc_auc_score(y_test, y_pred))
        
    average_importances = np.mean(feature_importances, axis=0)
    average_cross_val_score = np.mean(cross_val_scores)
    average_confusion_matrix = np.mean(confusion_matrices, axis=0)    
    average_roc_auc_score = np.mean(roc_auc_scores)
        
    return average_importances, average_cross_val_score, average_confusion_matrix, average_roc_auc_score, classification_reports
    
def print_result(result):
    average_importances = result[0]
    average_cross_val_score = result[1]
    average_confusion_matrix = result[2]
    average_roc_auc_score = result[3]
    classification_reports = result[4]
    
    tn, fp, fn, tp = average_confusion_matrix.ravel()
    
    print("Average accuracy:", average_cross_val_score)
    print("True Positives (TP):", tp)
    print("False Positives (FP):", fp)
    print("True Negative (TN):", tn)
    print("False Negative (FN):", fn)
    print("ROC AUC:", average_roc_auc_score)
    for i, report in enumerate(classification_reports):
        print(f"Iteration {i+1}:\n{report}")
```

```python
rf_classifier = RandomForestClassifier(n_estimators=150)

result = cross_validation_classifer(rf_classifier, X, y)
print_result(result)
```

```python
# plot Confusion Matrix
labels = ['rejected','welcomed']
sns.heatmap(result[2], annot=True,  cmap='Blues', xticklabels=labels, yticklabels=labels)
plt.title('Confusion Matrix: Random Froeset')
plt.xlabel('Predicted Label')
plt.ylabel('True Label')
plt.show()
```

```python
# plot Confusion Matrix

labels = ['welcomed', 'rejected']
tn, fp, fn, tp = result[2].ravel()
num_classes = len(labels)
x_ticks = np.arange(num_classes)
plt.bar(x_ticks, [tn, fn], color='#015AA2', label='Predicted as rejected')
plt.bar(x_ticks, [fp, tp], bottom=[tn, fn], color='#FFF200',label='Predicted as welcomed')
plt.xlabel('True Label')
plt.ylabel('Count')
plt.xticks(x_ticks, labels)
plt.legend()
plt.show()
```

```python
feature_importances_table = pd.DataFrame({'features': X.columns, 'importance': result[0]}).sort_values(by='importance', ascending=True).reset_index()
feature_importances_table
```
